```python
df_pop = pd.read_csv(
    "ine_population_data.csv",
    encoding="latin1",
    sep=";",
    skiprows=3
)

df_pop
```

```python
df_pop["year"] = df_pop["year"].ffill() #correct so all rows have an assigned year
df_pop
```

```python
# Temporarily show all rows
with pd.option_context('display.max_rows', None):
    display(df_pop) 
```

```python
df_pop = df_pop.iloc[0:73] #select the data of interest
df_pop
```

```python
#Instead of having region with the code and name, transform it into 2 columns
df_pop[["region_code", "region_name"]] = df_pop["region"].str.split(":", n=1, expand=True)
df_pop = df_pop.drop('region', axis=1) #delete region column
df_pop
```

```python
# checking null values
df_pop.info()
```

```python
#Since the column extra is all null values, we can delete it
df_pop = df_pop.drop('extra', axis=1)
```

```python
df_pop
```

```python
# Get all names of Area Metropolitan of Porto, municipalities
df_pop["region_name"] = df_pop["region_name"].str.strip() # Remove spaces in the begining and the end of the region names
unique_regions = df_pop["region_name"].dropna().unique().tolist()
display(unique_regions)
```

```python
df_pop.to_csv('cleaned_ine_population_data.csv')
```

This site has two weather stations in Porto, one in Serra do Pilar (Vila Nova de Gaia, Porto) and another one in Pedras Rubras (Maia, Porto).

Source: 
- Serra do Pilar: https://meteostat.net/pt/place/pt/porto?s=08546&t=2022-01-01/2026-02-28
- Pedras Rubras: https://meteostat.net/pt/place/pt/porto?s=08545&t=2022-01-01/2026-02-28#google_vignette

```python
df_station_sp = pd.read_csv("porto_serra_do_pilar_weather.csv")
df_station_pr= pd.read_csv("porto_pedras_rubras_weather.csv")
```

```python
print("Serra do Pilar")
display(df_station_sp)

print("Pedras Rubras")
display(df_station_pr)
```

```python
# We can delete the columns "snow", "wdir" and "tsun" because they have all values as null
df_station_sp = df_station_sp.drop(columns=["snow", "wdir", "tsun"])
df_station_pr = df_station_pr.drop(columns=["snow", "wdir", "tsun"])

print("Serra do Pilar")
display(df_station_sp)
print("Pedras Rubras")
display(df_station_pr)
```

# Charging Points Dataset


Number of Outlets at charging points by geographic location (NUTS - 2024) and outlet type; Monthly

Source: Instituto Nacional de Estatística (INE)

https://www.ine.pt/xportal/xmain?xpid=INE&xpgid=ine_indicadores&indOcorrCod=0014692&contexto=bd&selTab=tab2


```python
df_charging = pd.read_csv(
    "ine_charging_points_number_location.csv",
    encoding="latin1",
    sep=";",
    skiprows=3
    )

df_charging
```

```python
row_period = df_charging.iloc[1]
row_type = df_charging.iloc[3]

columns = []
current_period = None

for i in range(len(df_charging.columns)):
    if pd.notna(row_period[i]):
        current_period = row_period[i]
    
    socket_type = row_type[i]
    
    if i == 0:
        columns.append("region")
    elif pd.notna(socket_type):
        columns.append(f"{current_period} | {socket_type}")
    else:
        columns.append(None)

df_charging.columns = columns

df_charging
```

```python
# Remove header rows
df_charging = df_charging.iloc[5:].reset_index(drop=True)

# Remove empty columns
df_charging = df_charging.loc[:, df_charging.columns.notna()]

# Remove duplicate columns
df_charging = df_charging.loc[:, ~df_charging.columns.duplicated()]

df_charging = df_charging.replace("- -", pd.NA)

df_charging
```

```python
#Select data of interest
df_charging = df_charging.iloc[0:18]
df_charging
```

```python
df_charging[["code", "region"]] = df_charging["region"].str.split(": ", expand=True)
df_charging
```

```python
df_long = df_charging.melt(
    id_vars=["code", "region"],
    var_name="Period_Type",
    value_name="Value"
)
df_long
```

```python
df_long[["Period", "Outlet_Type"]] = df_long["Period_Type"].str.split(
    " | ", expand=True, regex=False
)
df_long
```

```python
df_long = df_long.drop('Period_Type', axis=1)
df_long
```

```python
# Extracting the year 
df_long["year"] = df_long["Period"].str.extract(r"(\d{4})").astype(int)

#Extracting the month
df_long["month_name"] = df_long["Period"].str.extract(r"^([A-Za-zçÇ]+)")

df_long
```

```python
# Map the Month_names into numbers
month_map = {
    "Janeiro": 1,
    "Fevereiro": 2,
    "Março": 3,
    "Abril": 4,
    "Maio": 5,
    "Junho": 6,
    "Julho": 7,
    "Agosto": 8,
    "Setembro": 9,
    "Outubro": 10,
    "Novembro": 11,
    "Dezembro": 12
}

df_long["month"] = df_long["month_name"].map(month_map)
df_long
```

```python
df_charging = df_long.drop(columns=["month_name"])
df_charging = df_charging.drop(columns=["Period"])

df_charging
```

```python
df_charging.to_csv('cleaned_ine_charging_points_number_location.csv')
```
